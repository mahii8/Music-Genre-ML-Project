```python
import pandas as pd

# This line reads the file you just uploaded
df = pd.read_csv('music_genre.csv')

# This line shows you the first 5 rows so you can see what the data looks like
df.head()
```

```python
import numpy as np

# Fix '?' entries in tempo and convert to float
df['tempo'] = pd.to_numeric(df['tempo'], errors='coerce')

# Remove any rows with missing values
df_clean = df.dropna()

# Remove non-numeric columns that won't work in the model
df_final = df_clean.drop(['instance_id', 'artist_name', 'track_name', 'obtained_date'], axis=1)

# Check remaining data and data types
df_final.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize encoder
le = LabelEncoder()

# Transform text categories to numeric labels
df_final['music_genre'] = le.fit_transform(df_final['music_genre'])
df_final['key'] = le.fit_transform(df_final['key'])
df_final['mode'] = le.fit_transform(df_final['mode'])

# Verify the first few rows
df_final.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Heatmap: Shows which features (like Loudness/Energy) move together
plt.figure(figsize=(10, 8))
sns.heatmap(df_final.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()

# 2. Boxplot: Shows how popularity varies across different genres
plt.figure(figsize=(12, 6))
sns.boxplot(x='music_genre', y='popularity', data=df_final)
plt.xticks(rotation=45)
plt.title('Popularity Distribution by Genre')
plt.show()

# 3. Scatter Plot: Visualizes how Energy and Acousticness separate genres
plt.figure(figsize=(10, 6))
sns.scatterplot(x='energy', y='acousticness', hue='music_genre', data=df_final, alpha=0.3)
plt.title('Energy vs Acousticness')
plt.show()
```
